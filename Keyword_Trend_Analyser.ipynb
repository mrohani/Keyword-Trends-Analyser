```
#@title
kw_list=['موبایل', 'لپ تاپ', 'تبلت','طلا','انتخابات']
pytrends.build_payload(kw_list, timeframe='today 3-m', geo='IR')
```

```
#@title
related_queries = pytrends.related_queries()
#related_queries.values()
```
